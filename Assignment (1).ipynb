Perform meaningful visualizations using the dataset. Bring at least 3 visualizations.

```python
import matplotlib.pyplot as plt
plt.boxplot(iris['SW'])
plt.title("Boxplot of SW")
plt.xlabel("SW")
plt.ylabel("Frequency")
plt.show()
```

```python
import seaborn as sns
```

```python
### Histogram representation of SL using seaborn
sns.histplot(x = iris['SL'])
plt.title("Distribution of SL")
plt.show()
```

```python
## Pairplots of numerical data of iris
num_iris=iris[["SL","SW","PL","PW"]]
```

```python
sns.pairplot(num_iris)
```

```python
## Violin plot of PL
sns.violinplot(x = iris['PL'])
plt.title("Violin plot of PL")
plt.show
```

```python
#3 Histogram representation of PW
iris['PW'].plot.hist()
plt.title("Distribution of PW")
plt.xlabel("PW")
plt.show()
```
